### 6. Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

```python
def cleanse_year(year_str):
    if "*" in year_str:
       year=year_str.replace("*","")
    else:
          year=year_str
    return int(year)

df["Census Year"]=df["Census Year"].apply(cleanse_year)

df
```

```python
def cleanse_pop(pop_str):
  if "," in pop_str:
       pop=pop_str.replace(",","")
  else:
          pop=pop_str
  return int(pop)

df["Total Population"]=df["Total Population"].apply(cleanse_pop)

df
```

```python

def cleanse_cost(cost_str):
    cost_str=cost_str.replace("$","")
    cost_str=cost_str.replace(",","")
    if " Billion" in cost_str:
        cost=float(cost_str.replace(" Billion",""))*1000000000
    else:
        cost=cost_str
    return int(cost)

df["Census Cost"]=df["Census Cost"].apply(cleanse_cost)

df
```

```python
def cleanse_avg_cost(avg_cost_str):
    avg_cost_str=avg_cost_str.replace("$","")
    avg_cost_str=avg_cost_str.replace(",","")
    if " cents" in avg_cost_str:
        avg_cost=float(avg_cost_str.replace(" cents",""))/100
    else:
        avg_cost=avg_cost_str
    return float(avg_cost)

df["Average Cost Per Person"]=df["Average Cost Per Person"].apply(cleanse_avg_cost)

df
```
